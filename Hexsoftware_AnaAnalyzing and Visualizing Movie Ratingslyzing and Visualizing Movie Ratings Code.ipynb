```python
print("Mean:", df['IMDB_Rating'].mean())
print("Median:", df['IMDB_Rating'].median())
print("Mode:", df['IMDB_Rating'].mode()[0])
```

```python
# Histogram
plt.figure(figsize=(10,6))

sns.set_style("whitegrid")

sns.histplot(
    df['IMDB_Rating'],
    bins=10,
    kde=True,
    color='purple'
)

plt.title(
    "Distribution of IMDB Ratings",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel(
    "IMDB Rating",
    fontsize=12
)

plt.ylabel(
    "Count",
    fontsize=12
)

plt.grid(alpha=0.3)

plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x=df['IMDB_Rating'])

plt.title("Box Plot of IMDB Ratings")

plt.show()
```

```python
#Top 10 Rated Movies
top_movies = df.sort_values(by='IMDB_Rating', ascending=False)

top_10 = top_movies[['Series_Title', 'IMDB_Rating']].head(10)

print(top_10)
```

```python
plt.figure(figsize=(12,6))

sns.set_style("whitegrid")

sns.barplot(
    x='IMDB_Rating',
    y='Series_Title',
    hue='Series_Title',
    data=top_10,
    palette='magma',
    legend=False
)

plt.title(
    "Top 10 Rated Movies",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel(
    "IMDB Rating",
    fontsize=12
)

plt.ylabel(
    "Movie Name",
    fontsize=12
)

plt.grid(alpha=0.3)

plt.show()
```

```python
# Top Genres
genre_counts = df['Genre'].value_counts().head(10)

print(genre_counts)
```

```python
plt.figure(figsize=(10,6))

genre_counts.head(10).plot(
    kind='bar',
    color=['orange','coral','gold','tomato','salmon']
)

plt.title("Top 10 Movie Genres", fontsize=16)
plt.xlabel("Genre")
plt.ylabel("Number of Movies")

plt.xticks(rotation=45)

plt.show()
```

```python
# Pair Plot
sns.pairplot(
    df[['IMDB_Rating', 'Meta_score', 'No_of_Votes']],
    diag_kind='kde'
)

plt.show()
```
